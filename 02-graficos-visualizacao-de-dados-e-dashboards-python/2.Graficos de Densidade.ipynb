```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando dados

base = pd.read_csv("../data/trees.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Histograma com a linha de distribuição de frequência da colune Height
#kde = linha de densidade

sns.histplot(base.iloc[:,1], kde=True, bins=6, color="blue").set(title="Arvores")
```

```python
#Linha de densidade da coluna Height

sns.kdeplot(base.iloc[:,1], color="blue").set(title="Arvore")
```
