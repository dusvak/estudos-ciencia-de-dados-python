```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importando dados

base = pd.read_csv("../data/trees.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Gráfico de dispersão, considerando o Volume e Circunferência
#Com Matplotlib

plt.plot(base.Girth, base.Volume)
plt.title("Arvores")
plt.ylabel("Volume")
plt.xlabel("Circunferencia")
```

```python
#Mesmo gráfico de dispersão, porém com marker invés de linha
#Com Matplotlib

plt.scatter(x=base.Girth, y=base.Volume, color="blue", facecolors = "none", marker = "*")
plt.title("Arvores")
plt.ylabel("Volume")
plt.xlabel("Circunferencia")
```

```python
#Mesmo gráfico de dispersão, porém com "afastamento" dos dados (jitter)
#Com Seaborn

sns.regplot(x=base.Girth,y=base.Volume, data = base, x_jitter = 0.3, fit_reg = False)

#jitter é para evitar sobreposição dos dados, para não ocuparem o mesmo espaço 
#fit_reg linha de tendencia, traçando uma médiana entre todos valores
```
