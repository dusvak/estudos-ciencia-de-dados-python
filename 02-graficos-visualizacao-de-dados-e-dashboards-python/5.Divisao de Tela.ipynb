```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando dados

base = pd.read_csv("../data/trees.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Circunferência com Volume

plt.scatter(base.Girth, base.Volume)
```

```python
#Circunferência com Altura

plt.scatter(base.Girth, base.Height)
```

```python
#Altura com Volume

plt.scatter(base.Height, base.Volume)
```

```python
#Histograma do Volume

plt.hist(base.Volume)
```

```python
#Unindo os Gráficos

#Criação de figura, no qual os gráficos serão posicionados
plt.figure(1)
plt.subplot(2,2,1)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume)
plt.subplot(2,2,4)
plt.hist(base.Volume)
```
