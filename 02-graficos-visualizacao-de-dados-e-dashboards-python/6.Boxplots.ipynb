```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando dados

base = pd.read_csv("../data/trees.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Visualizando dados em Boxplot

#patch_artist diz se o gr√°fico ser√° preenchido
#showfliers diz se haver√° pontos flutuantes de Outliers
#notch cria um entalho na mediana

plt.boxplot(base.Volume, vert = False, showfliers = False, 
            notch = True, patch_artist = True)
plt.title("arvores")
plt.xlabel("volume")
```

```python
#Visualizando dados por linha

plt.boxplot(base.T)
plt.title("arvores")
plt.xlabel("dados")
```

```python
#Visualizando de 3 boxplots no mesmo gr√°fico

plt.boxplot(base.Volume, vert = False)
plt.boxplot(base.Girth, vert = False)
plt.boxplot(base.Height, vert = False)
plt.title("arvores")
plt.xlabel("dados")
```
