```python
import pandas as pd
import seaborn as srn
```

```python
#Importando dados

base = pd.read_csv("../data/trees.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Visualiza√ß√£o de um boxplot
#orient = vertical ou horizontal

srn.boxplot(data=base.Volume, orient="v").set_title("Arvores")
```

```python
#Visualiza√ß√£o de v√°rios boxplots na mesma imagem

srn.boxplot(data = base)
```
