<font size="4">Boxplots com Seaborn</font>

```python
import pandas as pd
import seaborn as srn
```

```python
#Importando dados

base = pd.read_csv("../data/trees.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Visualização de um boxplot
#orient = vertical ou horizontal

srn.boxplot(data=base.Volume, orient="v").set_title("Arvores")
```

```python
#Visualização de vários boxplots na mesma imagem

srn.boxplot(data = base)
```
