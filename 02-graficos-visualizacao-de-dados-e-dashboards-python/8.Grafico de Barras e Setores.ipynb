```python
import pandas as pd
```

```python
#Importando dados

base = pd.read_csv("../data/insect.csv")
base.shape
```

```python
#Visualizando dados

base.head()
```

```python
#Agrupando contagem de uso por spray

agrupado = base.groupby(["spray"])["count"].sum()
agrupado
```

```python
#Gráfico de barras dos dados agrupados

agrupado.plot.bar(color = "gray")
```

```python
#Gráfico de pizza dos dados agrupados, com legenda

agrupado.plot.pie(legend = True)
```
