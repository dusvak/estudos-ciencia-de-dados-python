```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
#Importando dados

base = pd.read_csv("../data/chicken.csv")
```

```python
#Visualizando dados

base.head()
```

```python
#Criação de novo dataframe agrupando o atributo feed

agrupado = base.groupby(["feed"])["weight"].sum()
agrupado
```

```python
#Novo dataframe somente para testar os filtros do pandas

teste = base.loc[base["feed"] == "horsebean"]
teste
```

```python
#Histograma considerando somente o valor Horsebean

srn.histplot(base.loc[base["feed"] == "horsebean"].weight, kde=True).set(title="horsebean")
```

```python
#Histograma considerando somente o valor Casein

srn.histplot(base.loc[base["feed"] == "casein"].weight, kde=True).set(title="casein")
```

```python
#Histograma considerando somente o valor Linseed

srn.histplot(base.loc[base["feed"] == "linseed"].weight, kde=True).set(title="linseed")
```

```python
#Histograma considerando somente o valor Meatmeal

srn.histplot(base.loc[base["feed"] == "meatmeal"].weight, kde=True).set(title="meatmeal")
```

```python
#Histograma considerando somente o valor Soybean

srn.histplot(base.loc[base["feed"] == "soybean"].weight, kde=True).set(title="soybean")
```

```python
#Histograma considerando somente o valor Sunflower

srn.histplot(base.loc[base["feed"] == "sunflower"].weight, kde=True).set(title="sunflower")
```

```python
#Gráficos unidos

plt.figure()
plt.subplot(3,2,1)
srn.histplot(base.loc[base["feed"] == "horsebean"].weight, kde=True, bins=10).set(title="horsebean")
plt.subplot(3,2,2)
srn.histplot(base.loc[base["feed"] == "casein"].weight, kde=True, bins=10).set(title="casein")
plt.subplot(3,2,3)
srn.histplot(base.loc[base["feed"] == "linseed"].weight, kde=True, bins=10).set(title="linseed")
plt.subplot(3,2,4)
srn.histplot(base.loc[base["feed"] == "meatmeal"].weight, kde=True, bins=10).set(title="meatmeal")
plt.subplot(3,2,5)
srn.histplot(base.loc[base["feed"] == "soybean"].weight, kde=True, bins=10).set(title="soybean")
plt.subplot(3,2,6)
srn.histplot(base.loc[base["feed"] == "sunflower"].weight, kde=True, bins=10).set(title="sunflower")

#Ajusta o layout para não haver sobreposição
plt.tight_layout()
```
